# What are Machine Learning models?
A very superficial definition:

 * Machine Learning is a discipline that is broadly interested in the use of data to create automated systems, to 'program' computers.
 * The computers are programmed by giving them examples of input-output relationships (for example, input pictures of pieces in a production chain, with the output if there is a fabrication defect or not).
 * Then a computable function that maps from input to output is found (and automatic way of detecting defects from pictures).


 * How these functions are found is what is often considered a machine learning model.
 * Well known models are Neural Networks, which are inspired in how neurons process information in the brain.


*Machine learning is a large field and has other applications, such as dimensionality reduction or clustering), we are focusing on what is called 'supervised learning', which means that we have examples of input-ouput relationships available.*

# Why learn about Machine Learning models in a Discrete Choice class?

Machine Learning (ML) models provide benefits/useful tools along two main axes of application:

1. **Model specification:** Remember that in our logit models we had to specify which variables and what variable transformations should be used in the model. This was trial and error process or an 'art form' that depends on the data analysis knowledge or intuitions about the problem. (*There is no guarantee that they find them, but they are useful, especially when we have a lot of data*).

The main reason for needing ML models is related to point 1 above. These data types rarelly work as covariates if we input them directly into the model, they require special transformations. These transformations are often very complex, cannot be done manually.

# Neural networks: The perceptron

The perceptron is the most basic model that can be considered a Neural network, even though it is not a network.

We will start with a mathematical definition.

Imagine that we have a set of explanatory variables $x$, the perceptron computes the sign of the linear combination of the variables with some coefficients that are the parameters of the perceptron.

 * These paremeters are like the $\beta$ in our logit models. In machine learning they are often called weights and given the notation $W$.
 * The sign function is just a function that given a number, it returns -1 if the number is negative and 1 if the number is positive.

$$perceptron(x) = sign( \beta x ) $$


![single_layer_perceptron]( "Taken from https://www.tutorialspoint.com/tensorflow/tensorflow_single_layer_perceptron.htm")

Sometimes, the perceptron is defined with the **step function** instead of the $sign$ function. The step function ouptuts 0 is the input is 0 or a negative value, 1 if the input is positive.


*A clue: We can begin finding connections with the logit model, specifically the binary logit. Remember that the logit uses a linear combination an then applies the logistic transform. The perceptron uses a linear combination and then uses the sign or step transformation. If we use the step function version, we could say that a perceptron can be used for producng choice probabilities that are either 0 or 1 for the reference alernative.*

# Biological inspiration

The idea behind the perceptron (attributed to Frank Rosenblatt, 1958) is
inspired in how neurons work.

 * They receive some input electrical signal from elsewe (other neurons, special cells). The signal can come from one other neuron/cell or from many other (no signal is also a signal). 
 * Depending on the values of all the input signals, the neuron either 'activates' or not, meaning that it will generate an output electrical signal. This output signal can then be used as input in other neurons, or in other special cells.

In the mathematical approximation:
 * The input signals are the input or explanatory variables, the $x$.
 * The decision on whether the neuron activates or not is encoded by a linear combination. Each input signal is multiplied by a number, then we sume everything together. If the sign of the sum if positive, the neuron activates, if the sign is negative, the neuron does not activate.


![percep_biology.png]( "Picture taken from https://towardsdatascience.com/the-differences-between-artificial-and-biological-neural-networks-a8b46db828b7")

The basic perceptron is not connected to other 'neurons' so the output, whether it is activated or not is the final value that we are interested. We can do the analogy with choice modelling: The 'neuron' that is represented by the perceptron is the one in charge of deciding whether we want one alternative or the other.

If we connect the ouput of some perceptron to the input of other perceptrons we get a propert neural *network*. The nice property of the neural networks based on perceptrons is that they can compute any input-output function, so a neural network can theoretically represent any given input-output mapping.

This is part of the motivation for using these models, they are biologically inspired and can represent any mapping, including the one from the variables to choices, whithout having to find the transformations ourselves. *Note: There are other models with that property, for example, polynomials of high enough order can do that too, we have known that for hundreds of years.*

# The connection to the logit

The perceptron is a good theoretical model, but there is a problem, how do we estimate the model? How do we find good values for the $\beta$?

For a single perceptron this can be done, for a network it becomes more difficult (we can do it but it would take too much time).

A practical solution to that, which is also 'biologically inspired' is to use the sigmoid function as the transformation, instead of the step function.

As we see, they look relatively similar.
So by using the sigmoid function, we get **exactly** the binary logit model.
In this case, we got to the sigmoid by trying to approximate the step function by something more smooth so we can use calculus to find the values for the $\beta$. It is also biologically sound, because in nature many phenomena are often smooth, we can say that neurons do not go from '0 to 1' for an infinitesimal change in input, they gradually change, though the change can be very quick.

# Estimating the Neural Networks: Cross-entropy vs Maximum Likelihood

In ML, Neural networks are estimated by finding the $\beta$ values that minimize a measure called the cross-entropy, sort of the similarity between the distribution function of the distribution produced by the neural network and the target distribution the one in the examples.

Let's call the probabilities prodced by the network p_{ij} and the target probabilities $q_{ij}$, with $i$ identifying the observation in the sample and $j$ the class, we we have multiple output neural network.

Cross-entropy:
 $$ H(q,p) = - \sum_1^N q_{ij}\log(p_{ij})$$

In ML, we try to find values for the $\beta$ that minimize the value above.

Recall from the multinomial logit, we estimated it via maximum likelihood.
The likelihood function was the multinomial:

$$ likelihood= \prod_1^N p_{ij}^{y_{ij}} $$

With $p_{ij}$ are the choice probabilities produced by the model, and $y_{ij}$ the indicator of which alternative was chosen by the individual in the data.

Remember that for numerical reasons we maximize the loglikelihood instead of the likelihood.

$$ loglikelihood= \sum_1^N y_{ij}\log(p_{ij}) $$

In many cases, the $y_{ij}$ are the $q_{ij}$ are equivalent, in data, we often observed the choices, not the probabilities. But we can think of the $y_{ij}$ as probabilities, by assigning probability 1 to the observed alternative and 0 to the others. And this way we have linked neural networks as a special case of the Multinomial Logit, or vice-versa.

# When to use decision trees

* So far, decision trees are very good at dealing with the classic 'table' data, the same as used in tradictional logit models (salary, prices, age, etc.) as opposed to the complex variables types (pictures, text, etc).

* So even though they can automatically find variable transformations, in practice they do not work that well.

# Validation of ML models

* Because of the complexity (roughly speaking, number of parameters) ML models can fit the data arbitrarily well, so likelihood based comparisons become pointless.
* The way ML models are validated is through holdout validation, separating the dataset intro two parts, one is used only to estimate the model, the other is used to compute the validation measures (accuracy, confusion matrices, etc.)

# Interpretation of ML models

Because of the complex variable transformation, it is often said that ML models are not as interpretable as the classic methods in Discrete Choice.

Interpretation is often tied to the 'linearity' of the model. A model that is purely linear on the variables is very easy to interpret, the sign of the coefficients, willingness to pay, etc. everything becomes direct.
However, even in the multinomial logit, if we add a couple of transformations (polynomials, logarithms) and some interactions to a multinomial logit, we loose interpretability very quickly.

We can apply the same toolbox for interpretability/insights that we have been using. For example,
 * Notable what if scenarios such as Willingness to pay and other marginal effects, we can still use them, we just need to apply the definition. The problem is that we get a complex function instead of just a number.
 * The general what-if-scenarios are exactly the same!

# Interpretability through local linear approximations

* There are 'novel' techniques that try to get the best of both worlds, the approximation capacity of ML models and the interpretabiltiy of the simple linear models.

* The methodology is called Local interpretable model-agnostic explanations (LIME).

* The idea is to use an estimated ML model to the whole dataset and the approximate the ML model by a linear one, that is interpretabe, **but focusing on specifics parts of the domain of application**.

* For example we have a complex ML model trained for the whole population of public transpot users. It is very complex, we cannot interpret it. But we can try to find a linear model that mimics the complex model for a specific subset, for example, one model for the population over 50, another for the population below 50.
